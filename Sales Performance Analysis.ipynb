# Sales Performance

```python
# Create 'OrderYear' and 'OrderMonth' columns from 'OrderDate' column
df_sales['OrderDate'] = pd.to_datetime(df_sales['OrderDate'])

df_sales['OrderYear'] = df_sales['OrderDate'].dt.year
df_sales['OrderMonth'] = df_sales['OrderDate'].dt.month
```

As there are no missing values in all columns of the `df_sales` dataframe, we can easily calculate sales revenue and profit.

```python
# Create 'TotalSales' and 'Profit' columns
df_sales['TotalSales'] = df_sales['UnitPrice']*df_sales['SalesAmount']
df_sales['Profit'] = df_sales['TotalSales'] - (df_sales['TotalProductCost']*df_sales['SalesAmount']) - (df_sales['TaxAmount']*df_sales['SalesAmount'])
```

```python
# Sales revenue and profit by year
biz_performance_summary = df_sales.groupby('OrderYear').agg({'TotalSales':'sum', 'Profit':'sum'})
```

```python
biz_performance_summary.reset_index(inplace = True)
```

```python
biz_performance_summary
```

Look at the table above, total revenue in 2010 and 2014 are very small compared to other years. Let's consider the Company's business performance by month.

```python
biz_performance_summary_by_month = df_sales.groupby(['OrderYear','OrderMonth']).agg({'TotalSales':'sum', 'Profit':'sum'})
biz_performance_summary_by_month.reset_index(inplace = True)
```

Because sales data is only available for one month in each of the years 2010 and 2014, we should remove these 2 years out of our analysis.

```python
biz_performance_summary = biz_performance_summary.iloc[1:4]
```

```python
biz_performance_summary
```

```python
plt.bar(x = biz_performance_summary['OrderYear'], height = biz_performance_summary['TotalSales'])
plt.plot(biz_performance_summary['OrderYear'], biz_performance_summary['Profit'], color = 'red')
```

The Company's revenue halved in 2012 then increased by nearly 80% in the year after. Let's consider revenue by month in each year.

```python
biz_performance_summary_by_month = biz_performance_summary_by_month[biz_performance_summary_by_month['OrderYear'].isin(['2011','2012','2013'])]
```

```python
biz_performance_summary_by_month
```

```python
# Revenue in 2011
plt.bar(x = biz_performance_summary_by_month[biz_performance_summary_by_month['OrderYear'] == 2011]['OrderMonth'],
       height = biz_performance_summary_by_month[biz_performance_summary_by_month['OrderYear'] == 2011]['TotalSales'])
```

```python
# Revenue in 2012
plt.bar(x = biz_performance_summary_by_month[biz_performance_summary_by_month['OrderYear'] == 2012]['OrderMonth'],
       height = biz_performance_summary_by_month[biz_performance_summary_by_month['OrderYear'] == 2012]['TotalSales'])
```

In 2012, sales dropped in all 12 months, affecting the overall revenue.

```python
# Revenue in 2013
plt.bar(x = biz_performance_summary_by_month[biz_performance_summary_by_month['OrderYear'] == 2013]['OrderMonth'],
       height = biz_performance_summary_by_month[biz_performance_summary_by_month['OrderYear'] == 2013]['TotalSales'])
```

Until 2013, revenue followed an upward trend, reaching 3,421M in December.
